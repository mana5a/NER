```python
data=[]
for i in range (0, len(df)):
    obj=dict()
    obj['question']=str(df['question'][i])
    obj['answer']=str(df['answer'][i])
    data.append(obj)
# print(data)
```

```python
# data
len(data)
```

```python
tags = list()
for i in range(0, len(data)):
    #print(type(data[i]['question']))
    x = unicode(data[i]['question'], "utf-8")
    doc = nlp(x)
    l = list()
    for ent in doc.ents:
        l.append(ent.text)
        #rint(ent.text, ent.start_char, ent.end_char, ent.label_, i) 
    tags.append(l)
print(tags)
```
